```python
# Load the provided CSV file into a pandas DataFrame
import pandas as pd
acc_players_df = pd.read_csv('acc_players-2324F.csv', header=1)

# Display the first 5 rows
display(acc_players_df.head())
```

```python
# Calculate the total points scored by all players combined
total_points = acc_players_df['PTS'].sum()
print("Total points scored by all players combined:", total_points)
```

```python
# Find the player who has played the most minutes (MP)
most_minutes = acc_players_df.loc[acc_players_df['MP'].idxmax(), 'Player']
print("Player who has played the most minutes:", most_minutes)
```

```python
# Identify the top 5 players in terms of total rebounds (TRB)
top_5_rebounders = acc_players_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print("Top 5 players in terms of total rebounds:")
display(top_5_rebounders)
```

```python
# New DataFrame containing only players who played more than 500 minutes
players_500_minutes = acc_players_df[acc_players_df['MP'] > 500]
display(players_500_minutes.head())
```

```python
# Determine the player with the highest total assists (AST)
highest_assists_player = players_500_minutes.loc[players_500_minutes['AST'].idxmax(), 'Player']
print("Player with the highest total assists:", highest_assists_player)
```

```python
# Who are the top 3 Assist Leads in the League?
top_3_assist_leaders = players_500_minutes[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
print("Top 3 Assist Leaders in the League:")
display(top_3_assist_leaders)
```

```python
# Who are the top 3 Shot Blockers?
top_shot_blockers = players_500_minutes[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
print("Top 3 Shot Blockers in the League:")
display(top_shot_blockers)
```

```python
# Group the players by School and calculate the total points scored by each school
points_by_school = acc_players_df.groupby('School')['PTS'].sum().reset_index()
print("Total points scored by each school:")
display(points_by_school)
```

```python
# Group the players by School and calculate the total assists (AST) for each team
assists_by_school = acc_players_df.groupby('School')['AST'].sum().reset_index()
print("Total assists by each school:")
display(assists_by_school)
```

```python
# Sort the schools by total points scored and display the top 3 schools
points_by_school = acc_players_df.groupby('School')['PTS'].sum().reset_index()
top_3_schools = points_by_school.sort_values(by='PTS', ascending=False).head(3)
print("Top 3 schools by total points scored:")
display(top_3_schools)
```

```python
# Bonus: Create a bar chart showing the top 5 players by total points scored (PTS)
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
top_5_players_by_points = acc_players_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x='PTS', y='Player', data=top_5_players_by_points)
plt.title('Top 5 Players by Total Points')
plt.show()
```
