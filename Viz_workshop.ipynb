```python
import pandas as pd
gapminder_data = pd.read_csv('gapminder_dataset.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Life Expectancy
plt.figure(figsize=(10, 6))
sns.histplot(gapminder_data['Life Expectancy'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Life Expectancy', fontsize=16)
plt.xlabel('Life Expectancy', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Line plot of Life Expectancy over Time for top 5 populous countries
top_5_countries = gapminder_data.groupby('Country')['Population'].mean().nlargest(5).index

plt.figure(figsize=(14, 7))
for country in top_5_countries:
    country_data = gapminder_data[gapminder_data['Country'] == country]
    plt.plot(country_data['Year'], country_data['Life Expectancy'], marker='o', label=country)

plt.title('Life Expectancy Trends Over Time (Top 5 Populous Countries)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Life Expectancy', fontsize=12)
plt.legend(title='Country')
plt.show()
```

```python
import plotly.express as px
fig = px.scatter(
    gapminder_data,
    x='GDP per Capita',
    y='Life Expectancy',
    color='Country',
    title='GDP per Capita vs Life Expectancy',
    labels={'GDP per Capita': 'GDP per Capita', 'Life Expectancy': 'Life Expectancy'})
fig.show()
```
